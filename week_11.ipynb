```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.spatial import Voronoi, voronoi_plot_2d
import folium
from folium.plugins import HeatMap
```

```python
df= pd.read_csv('/content/BPD_Arrests(in).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df_male = df[df['Gender'] == 'M'].dropna(subset=['Latitude', 'Longitude'])
df_female = df[df['Gender'] == 'F'].dropna(subset=['Latitude', 'Longitude'])

map_center = [38.9072, -77.0369]
combined_map = folium.Map(location=map_center, zoom_start=11)

heat_map_male = df_male[['Latitude', 'Longitude']].values
heat_map_female = df_female[['Latitude', 'Longitude']].values

male_layer = folium.FeatureGroup(name="Male Heatmap")
female_layer = folium.FeatureGroup(name="Female Licenses Heatmap")

HeatMap(heat_map_male).add_to(male_layer)
HeatMap(heat_map_female).add_to(female_layer)

male_layer.add_to(combined_map)
female_layer.add_to(combined_map)

folium.LayerControl().add_to(combined_map)

combined_map_path = "Gender_combined_map_layers55.html"
combined_map.save(combined_map_path)

combined_map_path
```

The heat map with the overlap of the two genders visualizes that both female and male arrests are cluttered around the same area. There is more blue in the female heatmap which means that there are fewer arrests in females in certian areas. The heatmap is a great way to compare the data for both genders.
