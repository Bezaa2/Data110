```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

Task 1: Basic Bar Chart for Crime Rates in 2023

```python
crimes = ['', 'Murder', 'Rape (revised definition)', 'Robbery', 'Burglary', 'Aggravated Assault', 'Motor Vehicle Theft', 'Larceny']
rates = ['0', '5.7', '38.0', '66.5', '250.7', '264.1', '318.7', '1,347.2']


plt.figure(figsize=(20,5))
plt.bar(crimes, rates, label='crime rates', color='coral')
plt.title('Offenses per 100,000 People (2023)')
plt.xlabel('Crimes')
plt.ylabel('Rates')
plt.legend()
plt.grid(axis='y', linestyle='-', color='blue')
plt.show()
```

Task 2: Change in Crime Rates (2022 vs. 2023) with Unique Colors

```python
crimes = ['', 'Murder', 'Rape (revised definition)', 'Burglary', 'Larceny', 'Aggravated Assault', 'Robbery', 'Motor Vehicle Theft']
percentage_change = ['', '-12.0', '-9.8', '-8.1', '-4.9', '-3.3', '-0.8', '+12.0']
colors = ['white', 'red', 'black', 'green', 'orange', 'yellow', 'grey', 'blue']

plt.figure(figsize=(8, 5))
plt.title('Change in Crime Rates (2022 vs. 2023)')
plt.xlabel('Percentage change')
plt.ylabel('Crimes')
plt.barh(crimes, percentage_change, color=colors)
plt.plot()
```

Why does this bar graph make more sense than Task 1's bar graph?

This bar graph makes more sense because it is easier to see each crime with thier unique colors and it is easier to read the graph with the horizontal length.

# Combined Bar Chart for Crime Rates in 2022 and 2023

```python
crime_type = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape (revised definition)', 'Robbery']
offense_rates = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
Change_rate_Offenses = [-3.3, -8.1, -4.9, +12.0, -12.0, -9.8, -0.8]

plt.figure(figsize=(15, 10))
plt.bar(crime_type, offense_rates, label='2023')
plt.bar(crime_type, Change_rate_Offenses, label='2022 rate change', bottom=offense_rates)
plt.legend()
plt.xlabel('Offense type')
plt.ylabel('Change in offense rates')
plt.show()
```
