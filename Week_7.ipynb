```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

iris = sns.load_dataset('iris')

print(iris.head())
```

```python
iris.info()
```

```python
iris.describe()
```

#**1. Data Visualization**

## 1.1 Pair Plot

```python
sns.pairplot(iris, hue="species", palette="coolwarm")

plt.suptitle("Pairplot of all featurues in Iris", y=1)

plt.show()
```

## 1.2 Bar Plot

```python
sns.barplot(x="species", y="sepal_length", data=iris)

plt.title('Average Sepal Length for Each Iris Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')

plt.show()
```

## 1.3 Box Plot

```python
sns.boxplot(x="species", y="petal_width", data=iris, palette='coolwarm')

plt.title('Distribution of petal widths for each species')

plt.show()
```

## 1.4 Grouped Bar Plot

```python
species_length_grouped = iris.groupby('species').mean()[['sepal_length','petal_length']].reset_index()

species_length_grouped.plot(x='species', kind='bar')
plt.title("Average Sepal and Petal lengths by Species")
plt.ylabel("Average Length (cm)")
plt.show()
```

## 1.5 Violin Plot

```python
sns.violinplot(x="species", y="sepal_length", data=iris)

plt.title("Distribution of Sepal Length for Each Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")

plt.show()
```
