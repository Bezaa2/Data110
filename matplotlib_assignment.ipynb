2. Basic Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 4, 400)
plt.plot(x, np.sin(x))
plt.show()
y = x**2 - 2*x + 1
plt.plot(x, y)
plt.title("Quadratic Function")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

3. Multiple Lines

```python
x = np.linspace(.1, 4, 400)

f_x = x**3

g_x = np.exp(x)

h_x = np.log(x)


y_f = f_x
y_g = g_x
y_h = h_x


plt.plot(x, y_f, label='f_x = x^3', color='purple')
plt.plot(x, y_g, label='g(x) = e^x', color='yellow')
plt.plot(x, y_h, label='h(x) = log(x)', color='blue')


plt.title('Plot of f_x = x^3, g(x) = e^x, and h(x) = log(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

4. Styling Lines

```python
x = np.linspace(-2, 2, 100)
plt.plot(x, x**2, label='f_x = x^2', color='green')
plt.plot(x, x**3, label='g(x) = x^3', color='#FFDD44')
plt.plot(x, x**4, label='h(x) = x^4', color=(0.1, 0.2, 0.5))
plt.plot(x, x**5, label='i(x) = x^5', color='0.75')


plt.title('Plot of f_x = x^2, g(x) = x^3, h(x) = x^4, and i(x) = x^5')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

5. Axes Limits and Scaling

```python
x = np.linspace(-10, 10, 400)

f_x = 1 / (x**2 + 1)

y = f_x

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f_x = 1 / (x^2 + 1)', color='blue')
plt.title('Axes Limits')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(-10, 10)
plt.ylim(0, 1.1)
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f_x = 1 / (x^2 + 1)', color='green')
plt.title('Reversed Axes')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(10, -10)
plt.ylim(1.1, 0)
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f_x = 1 / (x^2 + 1)', color='red')
plt.title('Tight scaling')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axis('tight')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f_x = 1 / (x^2 + 1)', color='purple')
plt.title('Equal scaling')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()
```

6. Labels and Legends

```python
x = np.linspace(0, 10, 400)

f_x = np.sqrt(x)
g_x = x**(1/3)

plt.plot(x, f_x, label='$f(x) = x$', color='blue')
plt.plot(x, g_x, label='$g(x) = x^{1/3}$', color='red')

plt.title('Square root of f(x) = x and Cube root of g(x) = x^{1/3}')
plt.xlabel('x')
plt.ylabel('Function Values')
plt.legend()
plt.show()
```

7. Basic Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.subplot(1, 2, 1)
plt.plot(x, y, 'o', label='Random Points')
plt.title('Scatter Plot using plt.plot()')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x, y, label='Random Points')
plt.title('Scatter Plot using plt.scatter()')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

8. Customizing Scatter Plots

```python
x = np.linspace(0, 4 * np.pi, 100)

f_x = x * np.sin(x)
y = f_x

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o-', markersize=7, linewidth=2, markerfacecolor='Red', markeredgecolor='black', markeredgewidth=1.5)

plt.title('Customized Scatter Plot of $f(x) = x \sin(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

11. Style Comparison

```python
plt.style.use('seaborn-whitegrid')

x = np.linspace(0, 4 * np.pi, 100)

f_x = x * np.sin(x)
y = f_x

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o-', markersize=7, linewidth=2, markerfacecolor='Red', markeredgecolor='black', markeredgewidth=1.5)

plt.title('Customized Scatter Plot of $f(x) = x \sin(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
plt.style.use('ggplot')

x = np.linspace(0, 4 * np.pi, 100)

f_x = x * np.sin(x)
y = f_x

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o-', markersize=7, linewidth=2, markerfacecolor='Red', markeredgecolor='black', markeredgewidth=1.5)

plt.title('Customized Scatter Plot of $f(x) = x \sin(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
plt.style.use('dark_background')

x = np.linspace(0, 4 * np.pi, 100)

f_x = x * np.sin(x)
y = f_x

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o-', markersize=7, linewidth=2, markerfacecolor='Red', markeredgecolor='black', markeredgewidth=1.5)

plt.title('Customized Scatter Plot of $f(x) = x \sin(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

The diffrence between the 3 diffrent styles are the background.

The Seaborn library uses a white background with grey grid lines. It is a clean and professional look. It is great because its easy to distinguish data points and lines for a detailed data analysis and presentations where clarity is important.

The ggplot style is a clean and minimalist look with a grey background and white grid lines. This emphasizes the data points and lines. This is easy to read.

The darkbackground uses black background with white elements for the text and grid lines. It creates a high-contrast look this would be great for dark enviroment presntations.

For this data I think the most appropriate style is the Seaborn Library.
